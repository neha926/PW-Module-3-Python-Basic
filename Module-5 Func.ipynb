10. What is the difference between map(), reduce(), and filter() functions in Python?
 Function      Purpose                             Output           
  ----------  ----------------------------------  ---------------- 
  `map()`      Applies a function to all items     Transformed list 
  `filter()`   Keeps items that match a condition  Filtered list    
  `reduce()`   Combines all items into one value   Single value     


```python
from functools import reduce

nums = [1, 2, 3, 4]

map_result = list(map(lambda x: x*2, nums)) 
print(map_result)# [2, 4, 6, 8]
filter_result = list(filter(lambda x: x % 2 == 0, nums))
print(filter_result) # [2, 4]
reduce_result = reduce(lambda x, y: x + y, nums)    
print(reduce_result) # 10
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.


```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

is_prime(2)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.


```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

fib = FibonacciIterator(10)  # 10 terms
for num in fib:
    print(num)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.


```python
def read_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_lines("sample.txt"):
    print(line)
```

10. Create a Python program that uses filter() to remove all the vowels from a given string.


```python
def remove_vowels(s):
    return ''.join(filter(lambda ch: ch.lower() not in 'aeiou', s))

# Example: 
remove_vowels("Hello World")
```
